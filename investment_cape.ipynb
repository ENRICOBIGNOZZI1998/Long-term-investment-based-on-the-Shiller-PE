```python
ok.symbols_in_namespace('RATIO')
```

```python
chn = ok.Indicator('CHN_CAPE10.RATIO').values_monthly
ind = ok.Indicator('IND_CAPE10.RATIO').values_monthly
usa = ok.Indicator('USA_CAPE10.RATIO').values_monthly
euro=ok.Indicator('EUR_CAPE10.RATIO').values_monthly
```

```python
fig, ax = plt.subplots(figsize=[12, 6])
first_date_dt = pd.to_datetime('2006-01')  
last_date_dt = pd.to_datetime('2022-08')  

ax.plot(usa.index.to_timestamp(), usa, color ='green', linewidth = 2, label='usa')
ax.plot(ind.index.to_timestamp(), ind, color ='red', linewidth = 2, label='india')
ax.plot(chn.index.to_timestamp(), chn, color ='blue', linewidth = 2, label='cina')
ax.plot(chn.index.to_timestamp(), euro, color ='black', linewidth = 2, label='euro')

ax.legend(loc='upper right')
ax.set_xlim(first_date_dt, last_date_dt + pd.DateOffset(months=1))   
```
